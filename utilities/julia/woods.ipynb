```julia
using Plots
wood(a,b,c) = (t) ->  a * t^b * ^-(c*t)
w1 = wood(20,0.25,0.004) 
w2 = wood(30,0.15,0.004) 
w3 = wood(10,0.40,0.004)
plot([w1,w2, w3], 0,350, label = ["w1" "w2" "w3"])
```

In the plot above w1 has parameters similar to the overall means. The other two lactations appear to be similar except for a faster or slower rise after calving. Note that to achieve these shapes, I had to change **both** b and c parameters. These all look like credible lactations.

```julia
using Plots
wood(a,b,c) = (t) ->  a * t^b * ^-(c*t)
w1 = wood(20,0.25,0.005) 
w2 = wood(30,0.15,0.005) 
w3 = wood(10,0.40,0.005)
plot([w1,w2, w3], 0,350, label = ["w1" "w2" "w3"])
```

Here I increased the c parameter only. It does change the area under the curve, but it is relatively "efficient" in changing the shape more than the scale of the curve

```julia
using Plots
wood(a,b,c) = (t) ->  a * t^b * ^-(c*t)
w1 = wood(20,0.25,0.005) 
w2 = wood(20,0.15,0.005) 
w3 = wood(20,0.40,0.005)
plot([w1,w2, w3], 0,350, label = ["w1" "w2" "w3"])
```

Here I **only** change the b parameter to the values of the first graph. Note that there is a large "side effect" on the scale of production, if not the "scale" parameter.

```julia
using Plots
wood(a,b,c) = (t) ->  a * t^b * ^-(c*t)
aYear = [-.48,-.66,-.33,-.028,.13,.30,.16,.054,.34,.64]
bYear = [0.72, 1.00, 0.47, 0.24, -0.14, -0.58, -0.71, -0.51, 0.069, -0.12, -0.44]
cYear = [0.99, 1.29, 0.82, 0.45, 0.34, -0.36, -0.71, -0.57, -0.44, -0.78, -1.03]

aVar = map(a -> wood(19.9+a, 26.89/100, 33.76/10000), bYear )
bVar = map(b -> wood(19.9, (26.89 + b)/100, 33.76/10000), bYear )
cVar = map(c -> wood(19.9, 26.89/100, (33.76+c)/10000), bYear )

display(plot(aVar, 0,350, title = "Yearly effect of a parameter"))
display(plot(bVar, 0,350, title = "Yearly effect of b parameter"))
display(plot(cVar, 0,350, title = "Yearly effect of c parameter"))
```
